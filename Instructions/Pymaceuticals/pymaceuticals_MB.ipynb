## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np 

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_df = pd.DataFrame(mouse_metadata)
study_df = pd.DataFrame(study_results)
all_df = pd.merge(mouse_df, study_df, on="Mouse ID")
# Display the data table for preview
all_df
```

```python
# Checking the number of mice.
len(all_df['Mouse ID'].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup_ID = all_df[all_df.duplicated(["Mouse ID", "Timepoint"], keep = False)]
dup_ID = dup_ID.iloc[1,0]
dup_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_ID_df = all_df.loc[all_df["Mouse ID"] == 'g989', :]
dup_ID_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = all_df.drop_duplicates(subset=["Mouse ID", "Timepoint"], keep = 'last')
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_num = clean_df["Mouse ID"].value_counts()
clean_tot = len(clean_num)
clean_tot
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
clean_df.describe()
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
clean_drug = clean_df.groupby('Drug Regimen')

TVmean = clean_drug['Tumor Volume (mm3)'].mean()
TVmedian = clean_drug['Tumor Volume (mm3)'].median()
TVvar = clean_drug['Tumor Volume (mm3)'].var()
TVsdev = clean_drug['Tumor Volume (mm3)'].std()
TVsem = clean_drug['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary dataframe.
TV_stat = pd.DataFrame({"Mean Tumor Volume": TVmean, 
                       "Median Tumor Volume": TVmedian, 
                       "Tumor Volume Variance": TVvar, 
                       "Tumor Volume Std Dev": TVsdev, 
                       "Tumor Volume Std Error": TVsem})
TV_stat
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line
agg_TV = clean_drug['Tumor Volume (mm3)'].agg(['mean','median','var','std','sem'])
agg_TV_df = agg_TV.rename(columns={'mean':'Mean', 
                                  'median': "Median", 
                                  'var': 'Variance', 
                                  'std': 'Std Dev', 
                                  'sem': 'Std Error'})
agg_TV_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
cap_df = clean_df.loc[clean_df["Drug Regimen"]=="Capomulin"]
ram_df = clean_df.loc[clean_df["Drug Regimen"]=="Ramicane"]
inf_df = clean_df.loc[clean_df["Drug Regimen"]=="Infubinol"]
cef_df = clean_df.loc[clean_df["Drug Regimen"]=="Ceftamin"]

drugs = [cap_df, ram_df, inf_df, cef_df]
combined_df = pd.concat(drugs)
combined_df 
# Start by getting the last (greatest) timepoint for each mouse
mouse_df = clean_df.groupby("Mouse ID")
max_time = mouse_df["Timepoint"].max()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merged_df = pd.merge(combined_df, max_time, on ="Mouse ID")
merged_df["Max Time"] = merged_df["Timepoint_x"]== merged_df["Timepoint_y"]

#dropping values not equal to max_time
merged_df.drop(merged_df.loc[merged_df['Max Time'] ==False].index, inplace=True)

#new df with just the maximum value 
FTV_df = merged_df[["Mouse ID", "Drug Regimen", "Timepoint_x", "Tumor Volume (mm3)"]]
FTV_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatments: 
    vol = FTV_df.loc[FTV_df["Drug Regimen"]==drug]["Tumor Volume (mm3)"]
    tumor_vol.append(vol)
    # Locate the rows which contain mice on each drug and get the tumor volumes
    quartile = vol.quantile([0.25, 0.5, 0.75])
    lower = quartile[0.25]
    upper = quartile[0.75]
    iqr = upper - lower   
    
    # Determine outliers using upper and lower bounds
    low_bound = lower - (1.5*iqr)
    up_bound = upper + (1.5*iqr)
    outliers = vol.loc[(vol > up_bound) | (vol < low_bound)]
    if len(outliers) > 0: 
        print(f"The outliers for {drug} are {outliers}.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(tumor_vol, labels = treatments)
plt.xlabel("Drug Regimen")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Volumes per Drug Regimen")
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
#single mouse id is x401
mouse401 = cap_df.loc[cap_df["Mouse ID"] == "x401"]

#line plot
x = mouse401["Timepoint"]
y = mouse401["Tumor Volume (mm3)"]

#formatting
plt.plot(x,y,color = 'r', label = 'Mouse ID x401')
plt.title("Tumor Vol vs. Timepoint for Mouse x401")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
#finding averages
cap_per_mouse = cap_df.groupby("Mouse ID")
cap_weight_avg = cap_per_mouse["Weight (g)"].mean()
cap_TV_avg = cap_per_mouse["Tumor Volume (mm3)"].mean()

#making df 
avg_df = pd.DataFrame({"Weight (g)": cap_weight_avg, 
                      "Tumor Volume (mm3)": cap_TV_avg})
#scatter plot
x_val = avg_df["Weight (g)"]
y_val = avg_df["Tumor Volume (mm3)"]

plt.scatter(x_val, y_val, marker = "x", facecolors = 'c', edgecolors = "k", s = x_val, alpha = 0.5)
plt.title("Avg Tumor Vol vs Mouse Weight")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

#linear equation
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_val, y_val)
reg_val = x_val * slope + intercept 
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept, 2))

#scatter plot and format titles
plt.scatter(x_val, y_val, marker = 'x', facecolor = 'c', edgecolor = 'k', 
           s = x_val, alpha = 0.5)
plt.title("Tumor Vol and Weight")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")

#plot line 
plt.plot(x_val, reg_val, "r-")
plt.annotate(line_eq, (20,37), fontsize = 12, color = 'r')

#print correlation coeff 
print(f"The correlation coefficient between mouse weight and tumor volume {round(st.pearsonr(x_val, y_val)[0], 2)}")
```
